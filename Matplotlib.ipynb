```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df_loja = pd.DataFrame({
    'id':[33,213,23,1,5],
    "Produto": ["TV 32", "TV 55", "TV 50", "Soundbar", "Microondas"],
    "Quantidade": [10, 8, 12, 5, 7],
    "Preço Unitário": [1200, 2500, 1950, 450, 300],
    'data_da_venda': pd.to_datetime(['2023-01-01', '2023-03-02', '2023-02-03', '2023-01-04', '2023-01-05'])

})
```

```python
df_loja['Mes'] = pd.DatetimeIndex(df_loja['data_da_venda']).month
```

```python
df_loja_venda_mes = df_loja.groupby(['Produto', 'Mes'])['Quantidade'].sum()
```

```python
df_grouped = df_loja_venda_mes.unstack()
df_grouped.plot(figsize=(5,3), kind='line', marker='o')
plt.xticks([1,2,3], ['Jan', 'Fev', 'Mar'])
plt.ylabel('Quantidade Venda')
plt.title('Quantidade total vendida por mes')
plt.show()
```

```python
df_loja_venda_prod = df_loja.groupby('Produto')['Quantidade'].sum()
df_loja_venda_prod.plot(kind='pie', autopct='%1.1f%%')
plt.title('Qtd vendida por produto')
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.scatter(df_loja['Preço Unitário'], df_loja['Quantidade'], s=df_loja['Quantidade'] * 5,
            c=df_loja['Preço Unitário'],cmap='plasma')
plt.xlabel("preco unit")
plt.ylabel("qtd vendida")
plt.title('Preco x Quatidade vendida')
plt.colorbar(label='preco unit')
plt.show()
```
