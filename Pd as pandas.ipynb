```python
import pandas as pd
```

#criando dados


```python
dados_alunos = {'nome':['Alice', 'Ana', 'Maria', 'Luiza', 'Clara', 'Sofia', 'Davi'],
                'matematica':[75, 30, 85, 100, 60, 40,82],
                'ciencias':[20,45,90,60,100,40,35],
                'portugues':[45,70,90,50,80,50,75]
                }
```

```python
df_dados_alunos = pd.DataFrame(dados_alunos)
df_dados_alunos
```

```python
df_dados_alunos[['matematica', 'ciencias']]
```

#definir função mean()


```python
def cal_media(linha):
  return linha[['matematica', 'ciencias', 'portugues']].mean()


cal_media(df_dados_alunos)
```

```python
df_dados_alunos['media'] = df_dados_alunos.apply(cal_media, axis=1)
```

```python
df_dados_alunos
```
