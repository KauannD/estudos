#Frequência Proporcional em NLP

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
import matplotlib.pyplot as plt
import string

nltk.download('punkt_tab')
nltk.download('stopwords')
```

##Pré-processamento do Texto

```python
# texto
texto = """Com a evolução contínua da tecnologia, o processamento de linguagem natural (PLN) tornou-se uma ferramenta essencial na análise de grandes volumes de dados textuais. O PLN permite que máquinas compreendam e interpretem a linguagem humana, facilitando a extração de informações valiosas e a automação de tarefas que antes exigiam intervenção humana. Técnicas avançadas de PLN são utilizadas em diversas aplicações, desde assistentes virtuais até sistemas de recomendação, contribuindo significativamente para a eficiência e a personalização dos serviços oferecidos. À medida que o campo avança, novas metodologias e algoritmos são desenvolvidos para aprimorar a precisão e a eficácia do PLN, tornando-o uma área de estudo fascinante e em constante evolução."""

# tokenização
tokens = word_tokenize(texto.lower())

# remover pontuação e stopwords
stopwords_pt = set(stopwords.words('portuguese'))
tokens_limpos = [token for token in tokens if token not in stopwords_pt and token not in string.punctuation]

print("Tokens após pré-processamento:")
print(tokens_limpos)
```

##Cálculo da Frequência Proporcional

```python
# contagem de frequência absoluta
frequencia_absoluta = Counter(tokens_limpos)

# total de palavras no texto
total_palavras = len(tokens_limpos)

# cálculo da frequência proporcional
frequencia_proporcional = {palavra: freq / total_palavras for palavra, freq in frequencia_absoluta.items()}

# frequências
print("\nFrequência Absoluta:")
for palavra, freq in frequencia_absoluta.most_common(10):
    print(f"{palavra}: {freq}")

print("\nFrequência Proporcional:")
for palavra, freq in sorted(frequencia_proporcional.items(), key=lambda x: x[1], reverse=True)[:10]:
    print(f"{palavra}: {freq:.4f}")
```

##Visualização da Frequência Proporcional

```python
# palavras mais frequentes (proporcionalmente)
palavras, frequencias = zip(*sorted(frequencia_proporcional.items(), key=lambda x: x[1], reverse=True)[:10])

# gráfico
plt.figure(figsize=(10, 6))
plt.bar(palavras, frequencias, color='lightgreen')
plt.title("Frequência Proporcional das Palavras")
plt.xlabel("Palavras")
plt.ylabel("Frequência Proporcional")
plt.xticks(rotation=45)
plt.tight_layout()


plt.show()
```
