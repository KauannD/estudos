```python
import pandas as pd
import matplotlib as mp

```

```python
populacao_inicial = pd.DataFrame({
    'municipio': ['São Paulo', 'Rio de Janeiro', 'Salvador', 'Fortaleza', 'Brasília'],
    'populacao': [12182719, 6453682, 2886698, 2687420, 2974703]
})
populacao_inicial
```

```python
populacao_inicial.head(2)  # Mostra as duas primeiras linhas do DataFrame
populacao_inicial.tail(2)  # Mostra as duas últimas linhas do DataFrame
populacao_inicial.info()  # Mostra informações sobre o DataFrame, como o número de linhas e colunas, os tipos de dados das colunas e o uso de memória
populacao_inicial.describe()  # Mostra estatísticas descritivas do DataFrame, como média, mediana, desvio padrão, mínimo e máximo
populacao_inicial.sort_values(by='populacao', ascending=False)  # Ordena o DataFrame pela coluna 'populacao' em ordem decrescente
populacao_inicial.loc[2]  # Seleciona a linha com índice 2 do DataFrame
populacao_inicial.iloc[2]  # Seleciona a linha na terceira posição do DataFrame (índice 2)
populacao_inicial['municipio']  # Seleciona a coluna 'municipio' do DataFrame
populacao_inicial[['municipio', 'populacao']]  # Seleciona as colunas 'municipio' e 'populacao' do DataFrame
populacao_inicial[populacao_inicial['populacao'] > 2000000]  # Seleciona as linhas do DataFrame onde a coluna 'populacao' é maior que 2 milhões
populacao_inicial.groupby('municipio')['populacao'].sum()  # Agrupa o DataFrame pela coluna 'municipio' e calcula a soma da coluna 'populacao' para cada grupo
```

```python
dados = {'idade': [20, 30, 40, 50, 60],
        'altura': [1.90, 1.80, 1.90, 1.60, 1.75]}
dfdados = pd.DataFrame.from_dict(dados)
dfdados
```

## medidas


```python
print(dfdados.mean()) # media
print(dfdados['idade'].mean())
#mediana
print(dfdados['altura'].median())
#moda
print(dfdados['altura'].mode())
#variancia
print(dfdados['idade'].var())
#desvio padrao
print(dfdados['idade'].std())
#coeficiente de variação
coe = dfdados.idade.std() / dfdados.idade.mean() * 100 # escala percent
print(coe)
#medidas de forma
print(dfdados.idade.skew())
print(dfdados.altura.skew() * 100)
# curtose
print(dfdados.idade.kurtosis())

###MEDIDAS
dfdados.idade.describe()

##correlacao
dfdados.corr()
dfdados.corr(method='spearman')
```
