```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
import urllib
import csv
```

```python
# manual dos dados
# https://lib.stat.cmu.edu/datasets/boston
```

```python
boston_features = ['CRIM', 'ZN', 'INDUS', 'CHAS','NOX','RM', 'AGE','DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
```

```python

# DADOS_URL = "https://archive.ics.uci.edu/ml/machine-learning-database/housing/housing.data"
DADOS_URL = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
```

```python
# local_file_path, header= urllib.request.urlretrieve(DADOS_URL)
local_file_path = 'boston_housing.csv'
```

```python
boston = pd.read_csv(DADOS_URL)
```

```python
boston.to_csv('boston_housing.csv', index=False)
```

```python
boston = pd.read_csv(local_file_path)
```

```python
boston.describe()
boston.corr()
```

```python
X =boston.iloc[:,:-1]
```

```python
y =boston.iloc[:,-1]
```

```python
X_train , X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state=42)

modelo_gb = GradientBoostingRegressor()
```

```python
modelo_gb.fit(X_train, y_train)
```

```python
score = modelo_gb.score(X_test, y_test)
```

```python
print(f'score: {score}')
```

```python
dados_entrada = []

for feature in boston_features[:-1]:
  valor = float(input(f'Insira o valor para {feature}: '))
  dados_entrada.append(valor)

previsao = modelo_gb.predict([dados_entrada])
print(f'Previsao: {previsao}')
```

#COLETA DE DADOS E DATA STORYTELLING


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mascara = np.zeros_like(boston.corr())
triangle_indices = np.triu_indices_from(mascara)
mascara[triangle_indices] = True
plt.figure(figsize=(16,10))
sns.heatmap(boston.corr(), mask=mascara ,annot=True, annot_kws={'size':14})
sns.set_style('white')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

```python
sns.set()
sns.set_context('talk')
sns.set_style('white')
sns.jointplot(x=boston['dis'], y=boston['nox'], size=7, color='indigo', joint_kws={'alpha':0.5})
plt.show()
```

```python
sns.set()
sns.set_context('talk')
sns.set_style('whitegrid')
sns.jointplot(x=boston['tax'], y=boston['rad'], height=7, color='darkred', joint_kws={'alpha':0.5})
plt.show()
```

```python
sns.set()
sns.set_context('talk')
sns.set_style('whitegrid')
sns.jointplot(x=boston['crim'], y=boston['medv'], height=7, color='darkred', joint_kws={'alpha':0.5})
plt.show()
```
